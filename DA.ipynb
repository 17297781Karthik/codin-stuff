```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Define x and y
x = np.array([i for i in range(1, 100)])
y = np.cos(x)
z = np.tan(x)  # Using cosine for the z-axis

# Create a 3D figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the data in 3D
ax.plot(x, y, z)

# Set labels for the axes
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Show the plot
plt.show()
```

```python
from nltk import word_tokenize,sent_tokenize
```

```python
text="""" This is the first sentece. This is the second sentence. This is the third sentence.

"""
nltk.download('punkt')
```

```python
sents=sent_tokenize(text)
sents
```

```python
wds=[word for sent in sents for word in word_tokenize(sent) ]
wds
```

```python
def tokn(text):
  sents=sent_tokenize(text)
  wds=[word for sent in sents for word in word_tokenize(sent)]
  return wds
```

```python
tktext=tokn(text)
tktext
```

```python
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
sw=set(stopwords.words('english'))
```

```python
print(sw)
```

```python
def removestop(tktext):
  res=[]
  for word in tktext :
    if word not in sw:
      res.append(word)
  return res
```

```python
final=removestop(tktext)
print(len(tktext))
print(len(final))
```
