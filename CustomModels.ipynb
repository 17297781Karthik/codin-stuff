# **Custom Linear Regression model**



This is not a optimised model this was just coded to know how actually a model mathematically works in real...

```python
import numpy as np
import pandas as pd
```

```python
class LinearRegression:
  def __init__(self,learning_rate,itrs):
    self.learning_rate=learning_rate
    self.itrs=itrs

  def fit(self,X,Y):
    self.X=X
    self.Y=Y
    self.n,self.m=X.shape
    self.b=0
    self.wts=np.zeros(self.m)

    for i in range(self.itrs):
        self.update()

  def update(self):
    self.y_pred=self.predict(self.X)

    dw=-2*(self.X.T).dot(self.Y-self.y_pred)/self.n
    db=-2*np.sum((self.Y-self.y_pred))/self.n

    self.wts=self.wts-self.learning_rate*dw
    self.b=self.b-self.learning_rate*db


  def predict(self,X):
    return X.dot(self.wts.T)+self.b

  def eval(self):
    print(self.y_pred)
    print(self.accuracy(self.Y,self.y_pred))
    print(self.precision(self.Y,self.y_pred))




  def accuracy(self, y_true, y_pred):
        """Calculates the accuracy of the model."""
        correct_predictions = np.sum(y_true == y_pred)
        total_predictions = len(y_true)
        return correct_predictions / total_predictions

  def f1_score(self, y_true, y_pred):
        """Calculates the F1 score of the model."""
        precision = self.precision(y_true, y_pred)
        recall = self.recall(y_true, y_pred)
        if precision + recall == 0:
            return 0
        return 2 * (precision * recall) / (precision + recall)

  def precision(self, y_true, y_pred):
        """Calculates the precision of the model."""
        true_positives = np.sum((y_true == 1) & (y_pred == 1))
        predicted_positives = np.sum(y_pred == 1)
        if predicted_positives == 0:
            return 0
        return true_positives / predicted_positives

  def recall(self, y_true, y_pred):
        """Calculates the recall of the model."""
        true_positives = np.sum((y_true == 1) & (y_pred == 1))
        actual_positives = np.sum(y_true == 1)
        if actual_positives == 0:
            return 0
        return true_positives / actual_positives




```

```python



np.random.seed(0)
data = {
    'Feature1': np.random.rand(30),
    'Feature2': np.random.rand(30),
    'Target': np.random.rand(30)
}


df = pd.DataFrame(data)

X=df[['Feature1','Feature2']]
Y=df['Target']


df.head(2)


```

```python
model=LinearRegression(0.00001,5000)

model.fit(X,Y)


model.eval()
```
